**4) Buatlah status BMI untuk setiap siswa DS10 dari kategori berikut, lalu simpan
    data statusnya dengan nama kolom Status BMI.**
* Normal: 18.5-24.9
* Overweight: 25.0-29.9
* Obesity: 30.0-34.9
* Underweight: < 18.5

```python
Status_BMI = []
for elemen in df_DS10['BMI']:
    if elemen >= 18.5:
        if elemen <= 24.9:
            Status_BMI.append("Normal")
    if elemen >= 25.0:
        if elemen <= 29.9:
            Status_BMI.append("Overweight")
    if elemen >=30.0:
        if elemen <=34.9:
            Status_BMI.append("Obesity")
    if elemen <18.5:
        Status_BMI.append("Underweight")
    else: 
        False
         
```
